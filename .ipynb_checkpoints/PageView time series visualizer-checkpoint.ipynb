```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
# Import data (Make sure to parse dates. Consider setting index column to 'date'.)
df = pd.read_csv('fcc-forum-pageviews.csv', index_col = 'date')
```

```python
df.head()
```

```python
# Calculate threshold values for top 2.5% and bottom 2.5%
threshhold_high = df['value'].quantile(0.975)
threshhold_low = df['value'].quantile(0.025)

# Filter out rows where page views are outside the thresholds
df = df[(df['value'] < threshhold_high) & (df['value'] > threshhold_low)]
```

```python
df.plot(kind = 'line', title = 'Daily freeCodeCamp Forum Page Views 5/2016-12/2019', xlabel = 'Date', ylabel = 'Page Views')
```
